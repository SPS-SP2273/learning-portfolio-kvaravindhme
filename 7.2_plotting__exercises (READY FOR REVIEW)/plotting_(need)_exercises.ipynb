# Fill in the blanks

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# A simple mathematical plot

```python
import numpy as np

#generating x and y data
x = np.linspace(0, 50, 1000)
y = np.exp(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x', color='darkorange', size='15')
plt.ylabel('y', color='darkorange', size='15', rotation=0)
plt.title(r"$y = e^{(x/10)}\sin(x)$", size='20')
plt.grid(alpha=0.25)
plt.ylim([-1, 1])
```
